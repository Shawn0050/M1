```python
import numpy as np
import matplotlib.pyplot as plt

# 給定 U 的值
U = 50

# 定義一個 x 的範圍
for U in range(0,31,10):
    x = np.linspace(0, 2, 100)  # 避免x為0，因為y = U / (x + 1)
    # 解 y 作為 x 的函數
    y = U / (x + 1)
    # 繪製 U = xy + y
    plt.plot(x, y, label='U=XY+Y')
    plt.title('Utility Function U=xy+y')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)

for E in range(5,21,5):
    x = np.linspace(0, 2, 100)
    y = E - 10*x
    # 繪製 U = xy + y
    plt.plot(x, y, label=E, color='lightblue')
    plt.title('Utility Function U=xy+y')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)


plt.show()
```
