```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/shawn/Github/M1/金融計量/etf.csv')
df
```

```python
years = df['年']
months = df['月']

result = [f"{year}{month:02d}" for year, month in zip(years, months)]
result_df = pd.DataFrame(result, columns=['年月'])
print(result_df)
```

```python
df2 = pd.DataFrame()
df2['年月'] = result_df['年月']
df2['報酬率％_月']  = df['報酬率％_月']
df2['證券代碼']  = df['證券代碼']
df2
```

```python
result_df = df2.pivot(index='年月', columns='證券代碼', values='報酬率％_月')
result_df = result_df.rename_axis('date')
result_df = result_df[['0050 元大台灣50','0051 元大中型100','0052 富邦科技','0053 元大電子','0055 元大MSCI金融','0056 元大高股息','0057 富邦摩台','0061 元大寶滬深']]
result_df
```

```python
result_df.to_csv('/Users/shawn/Github/M1/金融計量/regression_data/etf_transform.csv')
```
