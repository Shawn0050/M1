```python
import numpy as np
import matplotlib.pyplot as plt

#  U 
U = 50

#  x 
for U in range(0,31,10):
    x = np.linspace(0, 2, 100)  # x0y = U / (x + 1)
    #  y  x 
    y = U / (x + 1)
    #  U = xy + y
    plt.plot(x, y, label='U=XY+Y')
    plt.title('Utility Function U=xy+y')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)

for E in range(5,21,5):
    x = np.linspace(0, 2, 100)
    y = E - 10*x
    #  U = xy + y
    plt.plot(x, y, label=E, color='lightblue')
    plt.title('Utility Function U=xy+y')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)


plt.show()
```
